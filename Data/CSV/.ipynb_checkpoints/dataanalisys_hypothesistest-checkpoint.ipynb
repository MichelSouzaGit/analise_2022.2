```python
# Análise Descritiva

import numpy as np  
import pandas as pd 
import seaborn as sns 
```

```python
df_bridge = pd.read_csv("test_iperf_com_driver-bridge.csv")
df_host = pd.read_csv("test_iperf_com_driver-host.csv")
```

```python
df_bridge_clone = df_bridge[:]
df_host_clone = df_host[:]
```

```python
# renomeando e separando as variáveis de interesse

df_bridge_clone = df_bridge_clone.rename(columns={'streams/0/seconds': 'TransmissionTime'})
df_bridge_clone = df_bridge_clone.rename(columns={'streams/0/bits_per_second': 'PerSecondsBits'})
df_bridge_clone = df_bridge_clone.rename(columns={'streams/0/rtt': 'Latency'})
bridge_TransmissionTime = df_bridge_clone['TransmissionTime']
bridge_PerSecondsBits = df_bridge_clone['PerSecondsBits']
bridge_Latency = df_bridge_clone['Latency']
```

```python
# renomeando e separando as variáveis de interesse

df_host_clone = df_host_clone.rename(columns={'streams/0/seconds': 'TransmissionTime'})
df_host_clone = df_host_clone.rename(columns={'streams/0/bits_per_second': 'PerSecondsBits'})
df_host_clone = df_host_clone.rename(columns={'streams/0/rtt': 'Latency'})
host_TransmissionTime = df_host_clone['TransmissionTime']
host_PerSecondsBits = df_host_clone['PerSecondsBits']
host_Latency = df_host_clone['Latency']
```

```python
# resumo de dados bridge
from statistics import median, mean, mode

# valores mínimos e máximos para as variáveis de interesse
minTransmissionTime = min(bridge_TransmissionTime)
maxTransmissionTime = max(bridge_TransmissionTime)
minPerSecondsBits = min(bridge_PerSecondsBits)
maxPerSecondsBits = max(bridge_PerSecondsBits)
minLatency = min(bridge_Latency)
maxLatency = max(bridge_Latency)

# somatórios das variáveis de interesse
sumTransmissionTime = sum(bridge_TransmissionTime)
sumPerSecondsBits = sum(bridge_PerSecondsBits)
sumLatency = sum(bridge_Latency)

# media, mediana e moda
meanTransmissionTime = mean(bridge_TransmissionTime)
medianTransmissionTime = median(bridge_TransmissionTime)
modeTransmissionTime = mode(bridge_TransmissionTime)

meanPerSecondsBits = mean(bridge_PerSecondsBits)
medianPerSecondsBits = median(bridge_PerSecondsBits)
modePerSecondsBits = mode(bridge_PerSecondsBits)

meanPerLatency = mean(bridge_Latency)
medianLatency = median(bridge_Latency)
modeLatency = mode(bridge_Latency)
```

```python
# resumo de dados host
from statistics import median, mean, mode

# valores mínimos e máximos para as variáveis de interesse 
minTransmissionTime = min(host_TransmissionTime)
maxTransmissionTime = max(host_TransmissionTime)
minPerSecondsBits = min(host_PerSecondsBits)
maxPerSecondsBits = max(host_PerSecondsBits)
minLatency = min(host_Latency)
maxLatency = max(host_Latency)

# somatórios das variáveis de interesse
sumTransmissionTime = sum(host_TransmissionTime)
sumPerSecondsBits = sum(host_PerSecondsBits)
sumLatency = sum(host_Latency)

# media, mediana e moda
meanTransmissionTime = mean(host_TransmissionTime)
medianTransmissionTime = median(host_TransmissionTime)
modeTransmissionTime = mode(host_TransmissionTime)

meanPerSecondsBits = mean(host_PerSecondsBits)
medianPerSecondsBits = median(host_PerSecondsBits)
modePerSecondsBits = mode(host_PerSecondsBits)

meanPerLatency = mean(host_Latency)
medianLatency = median(host_Latency)
modeLatency = mode(host_Latency)
```

```python
# Gráfico de Latência por Tempo de Transmissão Driver Bridge
import matplotlib.pyplot as plt

plt.plot(bridge_TransmissionTime, bridge_Latency)
plt.title('Latência por Tempo de Transmissão Driver Bridge')
plt.xlabel('Tempo de Transmissão')
plt.ylabel('Latência')
plt.show()
```

```python
# Gráfico de Latência por Tempo de Transmissão Driver Host
import matplotlib.pyplot as plt

plt.plot(host_TransmissionTime, host_Latency)
plt.title('Latência por Tempo de Transmissão - Driver Host')
plt.xlabel('Tempo de Transmissão')
plt.ylabel('Latência')
plt.show()
```

```python
# Teste de Hipótese

import numpy as np
import statsmodels.api as statmod
import matplotlib.pyplot as plt

# A variável "Latência" foi selecionada para o teste de hipótese
# Primeiro teste: os dados são paramétricos ou não paramétricos?

# Latência Driver Bridge
fig = statmod.qqplot(bridge_Latency, line='45')
plt.show()
```

```python
# Latência Driver Host
fig = statmod.qqplot(host_Latency, line='45')
plt.show()
```

```python
# De acordo com os resultados, ambas as distribuições de latência (Drivers Host e Bridge) +
# São distribuições não normais ou não paramétricas +
# Portanto, para o teste de hipótese, usaremos testes não paramétricos

# Para corroborar a escolha de testes paramétricos
from scipy.stats import shapiro

#Para o driver Host
shapiro(host_Latency)
```

```python
# Para o driver Bridge
shapiro(bridge_Latency)
```

```python
# Segunda hipótese, há diferença estatística relevante entre as medianas ?
from scipy.stats import mannwhitneyu

# Juntando as variáveis de latência de Driver Host e Driver Bridge em um novo DataFrame
ListLatency = zip(bridge_Latency, host_Latency)
DataFrameMW = pd.DataFrame(ListLatency, columns=['BridgeLatency', 'HostLatency'])

Result = mannwhitneyu(DataFrameMW['BridgeLatency'], DataFrameMW['HostLatency'])
```
